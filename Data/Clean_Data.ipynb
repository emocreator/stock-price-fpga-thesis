```python
import pandas as pd
import numpy as np
```

```python
# NSE
df = pd.read_csv('C:\\Users\\Daniel aghanya\\Pictures\\tiktok download\\Final_Year_Project-main\\Uni project\\NSE All Share Historical Data.csv')
df.dropna(inplace=True)
df = df.iloc[::-1].reset_index(drop=True)
df.head()
```

```python
def convert_volume(volume_str):
    if volume_str.endswith('M'):
        return float(volume_str.replace('M', '')) * 1e6
    elif volume_str.endswith('B'):
        return float(volume_str.replace('B', '')) * 1e9
    else:
        return float(volume_str)

# Apply the conversion function to the 'Vol.' column
df['Vol.'] = df['Vol.'].apply(convert_volume)
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Function to preprocess and convert percentage columns
def preprocess_df(df):
    for column in df.columns:
        # Check if the column's values are of type object (string) and contain a '%' sign
        if df[column].dtype == object and df[column].str.contains('%').any():
            # Remove '%' sign and convert to float
            df[column] = df[column].str.replace('%', '').astype(float) / 100
        # Additionally, remove commas from numeric columns that are still formatted as strings
        elif df[column].dtype == object and df[column].str.contains(',').any():
            df[column] = df[column].str.replace(',', '').astype(float)
```

```python
preprocess_df(df)
```

```python

df['Close'] = df['Price'].astype(np.float64)  # Assuming 'Close' is same as 'Price'
df['Date'] = pd.to_datetime(df['Date'])


# Rename columns to a more familiar format
df.rename(columns={
    'Open': 'open', 
    'High': 'high', 
    'Low': 'low', 
    'Close': 'close', 
    'Vol.': 'volume',
}, inplace=True)


df.head()
```

```python
# Calculate the required statistics for the 'Price' column
statistics = {
    'Average': df['Price'].mean(),
    'Median': df['Price'].median(),
    'Std Deviation': df['Price'].std(),
    'Minimum': df['Price'].min(),
    'Maximum': df['Price'].max(),
    '20th Percentile': df['Price'].quantile(0.2),
    '80th Percentile': df['Price'].quantile(0.8)
}

statistics
```

```python
# Convert the Date column to datetime type and extract the year
df['Year'] = pd.to_datetime(df['Date']).dt.year

# Group by year and calculate the required statistics for the Price column
stats_by_year_close = df.groupby('Year')['close'].agg([
    ('Average', 'mean'),
    ('Median', 'median'),
    ('Std Deviation', 'std'),
    ('Minimum', 'min'),
    ('Maximum', 'max'),
    ('20th Percentile', lambda x: x.quantile(0.2)),
    ('80th Percentile', lambda x: x.quantile(0.8))
])

stats_by_year_close.reset_index(inplace=True)
stats_by_year_close
```
